```python
#https://github.com/cellinebutuem/tarantino-cursing-visualization/blob/master/tarantino-visual.ipynb
print(data.movie.unique())
movies_map = {'Reservoir Dogs': 1992, 'Pulp Fiction': 1994,'Jackie Brown':1997, 'Kill Bill: Vol. 1':2003,
       'Kill Bill: Vol. 2':2004, 'Death Proof':2007, 'Inglorious Basterds':2009, 'Django':2012, 'Hateful Eight':2015,
       'Once Upon a Time':2019} # Year data from wikipedia
data['year'] = data.movie.map(movies_map)
```

```python
#Now that years have been inputed lets check if it can be seen by using the below function
#https://github.com/cellinebutuem/tarantino-cursing-visualization/blob/master/tarantino-visual.ipynb
data.sort_values(['year'], inplace=True)
data
```

```python
#Now the unique() function is used to find the unique elements of an array. Returns the sorted unique elements of an array.
#https://github.com/cellinebutuem/tarantino-cursing-visualization/blob/master/tarantino-visual.ipynb
types = data.type.unique()
words = data.word.unique()
print(types)
print(words)
print(data.word.value_counts()[data.word.value_counts() > 10])
```

A lot of the words have the same root word and a few are just the plural of another word on the list. To deal with these cases I'll count singular and plural of the same word as one word and I'll add boolean columns for each root word.

```python
#https://github.com/cellinebutuem/tarantino-cursing-visualization/blob/master/tarantino-visual.ipynb
replace_map = {'dicks':'dick', 'fucks':'fuck', 'motherfuckers':'motherfucker', 'bastards':'bastard', 
               'assholes':'asshole', 'asses':'ass', 'bitches':'bitch', 'fuckers':'fucker', 'gooks':'gook', 
               'cunts':'cunt', 'goddamned':'goddamn', 'damned':'damn'}
data.replace(replace_map, inplace=True)
data.word.unique()
```

Now we will begin to plot for more of a visual overview of what has happened. The plot below is just an overview of the swear words for each movie.

```python
#the below is used to remove warning box, when the cell is run bellow a warning of glyph would pop up and interfer with the output, so a warning filter is in place to prevent that.
#https://github.com/cellinebutuem/tarantino-cursing-visualization/blob/master/tarantino-visual.ipynb
import warnings
warnings.filterwarnings('ignore')

data.word.fillna('NaN', inplace=True)
data['fuck'] = ['fuck' in w for w in data.word]
data['ass'] = ['ass' in w for w in data.word]
data['shit'] = ['shit' in w for w in data.word]
racial = ['japs', 'jew (verb)', 'n-word ', 'negro']
data['racial'] = [w in racial for w in data.word]
data['swearing'] = [w == 'word' for w in data.type]
data['other']=data.word
for i in range(len(data.word)):
    data['other'][i] = not(data.fuck[i]+data.ass[i]+data.racial[i]+data.shit[i] )
    
data
```

```python
#https://github.com/cellinebutuem/tarantino-cursing-visualization/blob/master/tarantino-visual.ipynb
data.word.replace({'NaN':np.nan},inplace=True) # for plotting purposes
sns.set(style='whitegrid', context='notebook', palette='PiYG')
fig, m = plt.subplots(9, 1, figsize=(18, 20), sharex=False)
plt.subplots_adjust(hspace = 0.9)
movies = data.movie.unique()
x = [data.word[(data.movie == m)] for m in movies]
xa = data.word
for i in range(9):
    m[i] = sns.countplot(x=x[i], data=data, ax=m[i])
    m[i].set_xticklabels(m[i].get_xticklabels(), rotation=30)
    m[i].set_ylabel(movies[i])
    m[i].set_xlabel('')
```
